```python
columnas_corregidas=['Identificación', 'Profundidad (ft)', 'Profundidad (m)',
       'Presión de Confinamiento (psi)', 'Presión de Confinamiento (MPa)', 'Longitud (cm)', 'Diámetro (cm)',
       'Vol. Total (cm³)', 'Vol. Poroso (cm³)', 'Vol. de Granos (cm³)', 'Densidad de Granos (g/cm³)',
       'Porosidad (%)', 'Permeabilidad Klinkenberg (md)','Permeabilidad Klinkenberg (µm²)',
       'Permeabilidad al Aire (md)', 'Permeabilidad al Aire (µm²)', 'B', 'Hi', 'RQI', 'Factor b (psi)',
       'Alpha', 'Fecha'] # Modificamos el listado de las columnas para nuestro DataFrame con nuestros requerimientos

basicos_pozox.columns=columnas_corregidas #Actualizamos las columnas de nuestro DataFrame
basicos_pozox.head(5)
```

```python
#Eliminamos la fila cero y reorganizamos las filas para tener como fila cero la primer fila del DataFrame
basicos_pozox=basicos_pozox.drop([0],axis=0)
basicos_pozox=basicos_pozox.reset_index(drop=True)
basicos_pozox.head(5)
```

```python
#Para modificar el tipo de dato de cada columna utilizamos una tupla para asignar el tipo de dato a cada columna
basicos_pozox = basicos_pozox.astype({'Identificación':'object', 'Profundidad (ft)':'float64', 'Profundidad (m)':'float64',
       'Presión de Confinamiento (psi)':'int32', 'Presión de Confinamiento (MPa)':'float64',
       'Longitud (cm)':'float64', 'Diámetro (cm)':'float64', 'Vol. Total (cm³)':'float64',
       'Vol. Poroso (cm³)':'float64', 'Vol. de Granos (cm³)':'float64',
       'Densidad de Granos (g/cm³)':'float64', 'Porosidad (%)':'float64',
       'Permeabilidad Klinkenberg (md)':'float64', 'Permeabilidad Klinkenberg (µm²)':'float64',
       'Permeabilidad al Aire (md)':'float64', 'Permeabilidad al Aire (µm²)':'float64', 'B':'float64', 'Hi':'float64',
       'RQI':'float64', 'Factor b (psi)':'float64', 'Alpha':'float64', 'Fecha':'float64'})

```

```python
basicos_pozox['B']=basicos_pozox['B'].apply(lambda x: format(x,'.1E')) #aplicamos notación cientifica para la columna B


# Se utiliza un bucle while para dar ciertas reglas para cantidad de cifras decimales para la permeabilidad
i=0
teta=[]       
max_index = len(basicos_pozox['Identificación'])

while i < max_index:
    if basicos_pozox.iloc[i,12]>10:
        teta.append(int(basicos_pozox.iloc[i,12]))
#         teta.append(np.format_float_positional(np.float(basicos_pozox.iloc[i,12]), unique=False, precision=1))
        i=i+1
    elif basicos_pozox.iloc[i,12]<=1:
        teta.append(round(basicos_pozox.iloc[i,12],3))
        i=i+1
    else:
        teta.append(np.format_float_positional(np.float(basicos_pozox.iloc[i,12]), unique=True, precision=1))
        i=i+1


        
basicos_pozox['Permeabilidad Klinkenberg (md)']=teta
```

```python
# Condicional para Revisar si la medición del Volumen poroso corresponde al comportamiento normal 
# donde el Vp dismuye al aumentar la Presion de Confinamiento sobre la muestra; si no sigue este 
#comportamiento debido a alguna migración de finos en la muestra o un compotamiento atipico en el equipo.


muestras_a_revisar=[]
i=0
max_index=len(basicos_pozox['Identificación'])

while i<max_index-1:
    if basicos_pozox.iloc[i,1]==basicos_pozox.iloc[i+1,1]:
        if basicos_pozox.iloc[i,8]>basicos_pozox.iloc[i+1,8]:
            i=i+1
        else:
            muestras_a_revisar.append(basicos_pozox.iloc[i,1])
            print ("La muestra:", basicos_pozox.iloc[i,1],"debe repetirse debido a aumento en el Vp")
            i=i+1
    else:
        i=i+1


```

```python
basicos_pozox[basicos_pozox['Profundidad (ft)'].isin(muestras_a_revisar)] #Muestras a Repetir debido a Aumento del Vp
```

```python
# Condicional para Revisar si la medición de la Permeabilidad Klinkenberg corresponde al comportamiento normal 
# donde la K dismuye al aumentar la Presion de Confinamiento sobre la muestra; si no sigue este 
#comportamiento debido a alguna migración de finos en la muestra o un compotamiento atipico en el equipo.


muestras_a_revisark=[]
i=0
max_index=len(basicos_pozox['Identificación'])

while i<max_index-1:
    if basicos_pozox.iloc[i,1]==basicos_pozox.iloc[i+1,1]:
        if basicos_pozox.iloc[i,12]>basicos_pozox.iloc[i+1,12]:
            i=i+1
        else:
            muestras_a_revisark.append(basicos_pozox.iloc[i,1])
            print ("La muestra:", basicos_pozox.iloc[i,1],"debe repetirse debido a aumento en la Permeabilidad en el cambio de presión de confinamiento de",basicos_pozox.iloc[i,3],"a", basicos_pozox.iloc[i+1,3])
            i=i+1
    else:
        i=i+1
```

```python
basicos_pozox[basicos_pozox['Profundidad (ft)'].isin(muestras_a_revisark)] #Muestras a Repetir debido a Aumento de Permeabilidad
```
